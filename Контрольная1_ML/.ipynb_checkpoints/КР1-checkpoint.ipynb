```python
dfm = df
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
dfm[['Кількість відгуків в гуглі']] = imputer_mean.fit_transform(dfm[['Кількість відгуків в гуглі']]).round()
dfm[['Средний чек']] = imputer_mean.fit_transform(dfm[['Средний чек']]).round()
dfm[['Середній дохід населення']] = imputer_mean.fit_transform(dfm[['Середній дохід населення']]).round()
dfm[['Населення середнє']] = imputer_mean.fit_transform(dfm[['Населення середнє']]).round()
```

```python
# Log Transformations for Skewed Distributions
df[['Средний чек', 'Супермаркет']] = np.log(df[['Средний чек', 'Супермаркет']])

import seaborn as sns
sns.set()
sns.pairplot(df)
```

```python
#encoding
dfs = df[['Тип']]
info = []
for column in dfs.columns:
        vals = {"col": column,
                "nunique": dfs[column].nunique(),
                "unique": dfs[column].unique(),
                "type": dfs[column].dtype
               }
        info.append(vals)
pd.DataFrame(info).sort_values(by = "col", ascending = False)
```

```python
type = {'Light':0,'light':0,'Standart':1,'Micro':2,'Premium':3}
df['Тип'] = df['Тип'].map(type)
```

```python
df = df.drop(['Этаж','Супермаркет'], axis='columns', inplace=False)
df.head(5)
```

```python
# Cheking correlations
import seaborn as sns
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, annot=True)
plt.show()
```

```python
# Splitting the dataset into the Training set and Test set
X = df.iloc[:,0:3]
y = df['Количество чеков в месяц']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
sr = LinearRegression().fit(X_train, y_train)

# Getting parameters
sr.coef_, sr.intercept_
```

```python
# Predicting the Test set results
y_pred = sr.predict(X_test)

# Coefficient of determination R^2
sr.score(X_train, y_train).round(3)
```

```python
# Mean squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, sr.predict(X_train), squared=False).round(2), mean_squared_error(y_test, y_pred, squared=False).round(2)
```

```python
# Visualising the Test set results
plt.scatter(X_test['Количество сотрудников в штаті'], y_test, color = 'red')
plt.plot(X_test['Количество сотрудников в штаті'], sr.predict(X_test), 'bo')
plt.title('Количество чеков в месяц vs Количество сотрудников в штаті (Test set)')
plt.xlabel('Количество сотрудников в штаті')
plt.ylabel('Количество чеков в месяц')
plt.show()
```

```python
# Checking p-values
import statsmodels.api as sm
X1 = sm.add_constant(X_train)
mr1 = sm.OLS(y_train, X1).fit()
#mr1.pvalues
mr1.summary()
```

```python
#Polynomial Regression
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
X_train_p = PolynomialFeatures().fit_transform(X_train)
X_test_p = PolynomialFeatures().fit_transform(X_test)

pr = LinearRegression().fit(X_train_p, y_train)

# Getting parameters
pr.coef_, pr.intercept_
```

```python
# Predicting the Test set results
y_pred = pr.predict(X_test_p)

# Coefficient of determination R^2
pr.score(X_train_p, y_train).round(3)
```

```python
#R2 little bit better
```

```python
# Mean squared error
mean_squared_error(y_train, pr.predict(X_train_p), squared=False).round(2), mean_squared_error(y_test, y_pred, squared=False).round(2)
```
